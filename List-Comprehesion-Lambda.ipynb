19. Create a list of unique elements from a list using list comprehension + 
condition.

```python
l1 = [1, 2, 3, 2, 4, 1, 5, 3]

l2 = [n1 for i, n1 in enumerate(l1) if n1 not in l1[:i]]

print(l2)
```

```python
l1 = [1, 2, 3]   
l2 = ['a', 'b']  

l3 = [(n1, num) for n1 in l1 for num in l2]

print(l3)
```

## **Lambda functions**

1. Write a lambda to add two numbers. 

```python
add = lambda n1, n2: n1 + n2
print(add(5, 3))
```

2. Create a lambda to check if a number is even. 




```python

is_even = lambda n1: "Even" if n1 % 2 == 0 else "Odd"

print(is_even(10))
print(is_even(7))
```

3. Write a lambda to get the last character of a string. 

```python
last_char = lambda s1: s1[-1]
print(last_char("Python"))
```

4. Use lambda with map() to square every number in a list. 

```python
l1 = [1, 2, 3, 4, 5]
l2 = list(map(lambda n1: n1 ** 2, l1))
print(l2)
```

5. Use lambda with filter() to get only odd numbers from a list. 

```python
l1 = [1, 2, 3, 4, 5, 6, 7]
l2 = list(filter(lambda n1: n1 % 2 != 0, l1))
print(l2)
```

6. Use sorted() + lambda to sort a list of tuples by second value. 

```python
l1 = [(1, 3), (4, 1), (2, 5), (6, 2)]
l2 = sorted(l1, key=lambda n1: n1[1])
print(l2)
```

7. Create a lambda to check if a string is a palindrome. 

```python
is_palindrome = lambda s1: s1 == s1[::-1]

print(is_palindrome("madam"))
print(is_palindrome("python"))
```

8. Use lambda to find maximum of three numbers. 

```python
max = lambda n1, n2, n3: n1 if n1 >= n2 and n1 >= n3 else (n2 if n2 >= n3 else n3)

print(max(10, 25, 15))
```

9. Write a lambda to reverse a string.

12. Use lambda to compute factorial using reduce() (yeah, that one-liner 
madness). 

```python
from functools import reduce

fact = lambda n1: reduce(lambda x, y: x * y, range(1, n1 + 1))

print(fact(5))
```
